```python
import pandas as pd
import plotly.graph_objs as go 
```

```python
df = pd.read_csv("F:/03. Suresh/1. Material/05. Data Science/20. Internship/1. Duit/1. Stocks Sept 2020/stocks_data_2020_till_2020_09_03.csv", skiprows=900000, nrows=1037084)
df.columns = ['id','timestamp','open_price','high_price','low_price','close_price','volume','scrip_id']
df.head()
```

```python
df.tail()
```

```python
df = df[:100000]
```

```python
# set the index
df = df.set_index(pd.DatetimeIndex(df['timestamp'].values))
df.head()
```

```python
figure = go.Figure(
         data = [
             go.Candlestick(
             x = df.index,
             low = df['low_price'],
             high = df['high_price'],
             close = df['close_price'],
             open = df['open_price'],
             increasing_line_color = 'green',
             decreasing_line_color = 'red'
             )
         ]
)

figure.show()
```

```python
figure = go.Figure(
         data = [
             go.Candlestick(
             x = df.index,
             low = df['low_price'],
             high = df['high_price'],
             close = df['close_price'],
             open = df['open_price'],
             increasing_line_color = 'green',
             decreasing_line_color = 'red'
             )
         ]
)

figure.update_layout(xaxis_rangeslider_visible = False)

figure.show()
```
